```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 500
std = 0.6
```

```python
balloons = pd.DataFrame({
    'length': np.random.normal(6, std, N),
    'width': np.random.normal(6, std, N),
    'weight': np.random.normal(2, std, N),
    'class': np.zeros(N, dtype='int')}
)

melons = pd.DataFrame({
    'length': np.random.normal(6, std, N),
    'width': np.random.normal(6, std, N),
    'weight': np.random.normal(10, std, N),
    'class': np.ones(N, dtype='int')}
)

fruits = pd.concat([balloons, melons])
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter3D(balloons['length'], balloons['width'], balloons['weight'], color='red', label='balloons')
ax.scatter3D(melons['length'], melons['width'], melons['weight'], color='orange', label='melons')
ax.legend()
ax.set_xlabel('Length')
ax.set_ylabel('Width')
ax.set_zlabel('weight')
plt.show()
```
