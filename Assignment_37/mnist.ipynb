```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
mnist_img = cv2.imread('input/mnist.png', 0)
plt.imshow(mnist_img, cmap='gray')
```

```python
mnist_img.shape
```

```python
rows, cols = mnist_img.shape
mnist_number = [None for i in range(10)]
for i in range(0, rows, 100):
    mnist_number[i//100] = mnist_img[i:i+100, 0:cols]
    cv2.imwrite(f'output/number_{i//100}.jpg', mnist_number[i//100])
    plt.subplot(10, 1, i//100 + 1)
    plt.imshow(mnist_number[i//100], cmap='gray')
```

```python
numbers = [None for i in range(500)]
rows, cols, steps = 100, 2000, 20

for i in range(len(mnist_number)):
    n = 1
    for j in range(0, rows, steps):
        for k in range(0, cols, steps):
            numbers[i] = mnist_number[i][j:j+20, k:k+20]
            cv2.imwrite(f'output/{i}/num{i}_{n}.jpg', numbers[i])
            n += 1
```
