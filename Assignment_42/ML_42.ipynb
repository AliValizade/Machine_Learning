```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

from modules import KNN

```

```python
female_data = pd.read_csv("data/ANSUR II FEMALE Public.csv")
female_data.head()
```

```python
male_data = pd.read_csv("data/ANSUR II MALE Public.csv", encoding="latin-1")
male_data.head()
```

```python
data = pd.concat([female_data, male_data])
data
```

# PreProccess

```python
data['weightkg'] = data['weightkg'] / 10    # Convert to kg
data['stature'] = data['stature'] / 10      # Convert to cm
data['Gender'] = data['Gender'].replace(['Female', 'Male'], [0, 1])     # Convert to binary
```

```python
cmap = ['red', 'blue']
plt.scatter(data['stature'], data['weightkg'], c=data['Gender'], cmap=ListedColormap(cmap), alpha=0.1)
plt.xlabel('Stature (cm)')
plt.ylabel('weight (kg)')
plt.title('Distribution based on Stature and weight')
plt.grid(True)
plt.show()
```

```python
f_data = data[data['Gender'] == 0]
m_data = data[data['Gender'] == 1]
```

```python
plt.hist(f_data['stature'], color='red', label='Female')
plt.hist(m_data['stature'], alpha=0.7, color='blue', label='Male')

plt.xlabel('Stature (cm)')
plt.ylabel('Count')
plt.title('Stature Distribution by Gender')
plt.legend()
plt.grid(True)

plt.show()
```

```python
X = data[['stature', 'weightkg']].values
Y = data['Gender'].values

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

print("Size of training set:", len(X_train))
print("Size of testing set:", len(X_test))
```

```python
k = [3, 5, 7, 9]
```

## Training and testing by my k-nearest-neighbors algorithm

```python
accuracies = []
for i in k:
    knn = KNN(i)
    knn.fit(X_train, Y_train)
    accuracies.append(knn.evaluate(X_test, Y_test))

results_df = pd.DataFrame({
    'k': k,
    'Accuracy': accuracies
})

results_df
```

```python
Y_pred = knn.predict(X_test)
```

```python
CM = np.zeros((2,2))

for i in range(int((len(data[["Gender"]])/100)*20)):
    if Y_test[i] == 0 and Y_pred[i] == 0:
        CM[0][0] += 1
    elif Y_test[i] == 0 and Y_pred[i] == 1:
        CM[0][1] += 1
    elif Y_test[i] == 1 and Y_pred[i] == 0:
        CM[1][0] += 1
    elif Y_test[i] == 1 and Y_pred[i] == 1:
        CM[1][1] += 1
    
CM
```

```python
plt.imshow(CM, interpolation='nearest', cmap=plt.cm.Wistia)
plt.colorbar()

classes = ['Female', 'Male']
tick_marks = np.arange(len(classes))
plt.xticks(tick_marks, classes)
plt.yticks(tick_marks, classes)

thresh = CM.max() / 2.
for i in range(CM.shape[0]):
    for j in range(CM.shape[1]):
        plt.text(j, i, int(CM[i, j]),
                 horizontalalignment="center",
                 color="white" if CM[i, j] > thresh else "black")

plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.title('Confusion Matrix')

plt.show()
```

## Training and testing by KNN algorithm with Scikit-learn

```python
accuracies = []
for i in k:
    knn = KNeighborsClassifier(i)
    knn.fit(X_train, Y_train)
    accuracies.append(knn.score(X_test, Y_test))

results_df = pd.DataFrame({
    'k': k,
    'Accuracy': accuracies
})

results_df
```

```python
Y_pred = knn.predict(X_test)
```

```python
cm = confusion_matrix(Y_test, Y_pred)
cm_display = ConfusionMatrixDisplay(confusion_matrix = cm, display_labels = ['Female', 'Male'])
cm_display.plot()
plt.show()
```
